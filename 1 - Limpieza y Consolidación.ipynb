## Detectar valores erróneos 

```python
# columnas con contenido
content_cols = ['lObjetivo', 'Estrategia', 'Nombre Actividad', 'Descripción del Plan', 'Nombre Actividad norm']

# selección manual de palabras cortas admisibles
palabras_cortas = ['BUZOS', 'E.IB.', 'CRA', 'PEI', 'PME','DUA', 'ADECO','JEC','PISE ','CRA ','PACE','Plan',
                   'ACLES', 'FODA','Viaje', 'Orden', 'Modem', '“PSU”','P.M.E.', 'WI-FI','Payas','SNCAE',
                   'Mapeo','HCP', 'UTP', 'Bono', 'Roles', 'PIUKE','RICE', 'TIC ','Mi pc.','Simce', 'Palin', 
                   'G.A.E', 'ATE','ACFIS', 'Aeles', 'LIRMI', ' PEI', 'ARPA', 'Kiwa', 'HPV', 'NEE', 'Pace', 
                   'PAI', 'Redes', 'ECAE', 'Tics', 'COSAM', 'Metas', 'CEDE','CREAR', 'Apoyo', 'CGA', 'local', 
                   'CAE', 'CEAL', 'flyer', 'CPEIP', 'acle', 'PACI', 'TICS','Jugar',  'Actas', 'Rifa','FORO', 
                   'AREAS', 'Rampa','Bingo','CORO','RICE', 'CRA', 'Pie', 'Aptus.','OA', 'Madre', 'CEAL', 'email', 
                   'yoga ', 'BECAS', 'Blog', 'Zoom', 'covid', 'cajas', 'P.A.P.P.I.','I.D.S.P', 'CANON', 'GTP','ABP',
                   'MINGA', 'TENS', 'CISCO', 'Pulso', 'Radar', 'SENDA','Plano', 'Gala', 'Bono','EDA', 'UAP','Mural',
                   'Datos', 'R.R.H.H''PAES', 'DÍA','PAL']
# normalización
palabras_cortas_norm = [re.sub(r'[^\w\s]', '', x) for x in palabras_cortas]
palabras_cortas_norm = [x.lower() for x in palabras_cortas_norm]
palabras_cortas_norm = [x.replace(' ','') for x in palabras_cortas_norm]

list_values_to_remove = []
for a in range(agno_inicial,agno_final+1):  
    print(a)
    data = pd.concat([planif_anual_dict[a][content_cols],reporte_implement_dict[a][content_cols]])
    for col in content_cols:
        j = 0
        nulls_index = []
        candidates0 = data[col].unique()
        candidates0 = [str(x) for x in candidates0 if len(str(x))<150]
        candidates = [re.sub(r'[^\w\s]', '', x) for x in candidates0]
        for text in candidates:
            # reglas de identificación de respuestas MALAS
            # 1: que use menos de 4 caracteres diferentes
            # 2: que tenga menos de 6 caracteres en total
            # 3: que sea numérico
            # 4: que no esté en lista de palabras cortas admisibles
            if ((len(set(text))<4)|(len(text)<6)|(text.isnumeric())) & (text.replace(' ','').lower() not in palabras_cortas_norm):
                nulls_index += [j]
            j+=1 
        list_values_to_remove += list(itemgetter(*nulls_index)(candidates0))
```

## Marcar valores erroneos como inválidos  -  consolidar  - generar llave

```python
def marcar_duplicados_nombre_accion(df):
    """Función que intenta recuperar nombre de acción desde descripción, de lo contrario bota filas"""
    # df no duplicados
    non_dup_df = df[~df.duplicated(['agno','rbd','Nombre Actividad norm'], keep=False)].copy()
    non_dup_df.loc[:,'es_duplicado_nombre_act'] = 0
    # df duplicados
    dup_df = df[df.duplicated(['agno','rbd','Nombre Actividad norm'], keep=False)].copy()
    dup_df.loc[:,'Nombre Actividad'] = dup_df['Descripción del Plan'].astype(str)
    dup_df.loc[:,'Nombre Actividad norm'] = \
    dup_df['Nombre Actividad'].apply(lambda x: normalizar_texto(x)) 
    # marcar duplicados
    dup_df.loc[:,'es_duplicado_nombre_act'] = 1
    # concatenar 
    df2 = pd.concat([non_dup_df, dup_df])
    # marcar duplicados post concatenación
    df2['es_duplicado_nombre_act2'] = df2.duplicated(['agno','rbd','Nombre Actividad norm'],keep=False).astype(int) \
                                      - df2.es_duplicado_nombre_act
    
    df2['fila_duplicada'] = df2.es_duplicado + (df2.es_duplicado_nombre_act + df2.es_duplicado_nombre_act2)
    
    df2 = df2.drop(columns = ['es_duplicado','es_duplicado_nombre_act','es_duplicado_nombre_act2'])

    return df2
```

```python
# Eliminar valores erroneos, tratar duplicados y consolidar

# Columnas llave
cols_match = ['agno', 'rbd','Nombre Actividad norm']

# Columnas en cada bbdd
cols_duplicadas = ['region', 'deprov', 'comuna', 'dependencia','Nombre Actividad', 'Dimensión',
                   'Subdimensión', 'lObjetivo', 'Estrategia', 
                   'Descripción del Plan', 'Fecha Inicio', 'Fecha Término',
                   'Programa Asociado', 'Nombre Responsable', 'Ate', 'planes']

# Para guardar filas duplicadas previo a merge anual
dict_duplicados_pre = {}
# Para guardar filas duplicadas post a merge anual
dict_duplicados_post = {}

# Para cada año
for a in agno_intervalo:  
    print(a)
    
    # recuperar cada bbdd
    data_plani = planif_anual_dict[a]
    data_impl = reporte_implement_dict[a]
    
    # reemplazar valores erróneos por 'valor_invalido'
    for col in content_cols + ['Nombre Actividad norm']:
        data_plani[col].replace(list_values_to_remove,'valor_invalido',inplace=True)
        data_impl[col].replace(list_values_to_remove,'valor_invalido',inplace=True)
        
    # Tratar duplicados según llave cols_match
    data_plani = marcar_duplicados_nombre_accion(data_plani)
    data_impl = marcar_duplicados_nombre_accion(data_impl)
    
    # Guardar en diccionarios filas duplicadas 
    dict_duplicados_pre['duplicados_planificacion{}'.format(a)] = data_plani[data_plani.fila_duplicada>0].copy()
    dict_duplicados_pre['duplicados_implementacion{}'.format(a)] = data_impl[data_impl.fila_duplicada>0].copy()
    
    # Quedarse sólo con una de las duplicadas 
    data_plani = data_plani.drop_duplicates(keep='first')
    data_plani = data_plani.drop_duplicates(subset = ['agno', 'rbd', 'Nombre Actividad norm'], keep='first') 

    data_impl = data_impl.drop_duplicates(keep='first')
    data_impl = data_impl.drop_duplicates(subset = ['agno', 'rbd', 'Nombre Actividad norm'], keep='first') 

    ## Consolidación planificación e implementación año "a" 
    # OBS: primero a tercero a continuación se hace así para mantener máxima información de columnas repetidas
    
    # primero: sin acciones "sólo de implementación" (left merge)
    data = pd.merge(data_plani,data_impl.drop(columns = cols_duplicadas), on = cols_match, how = 'left', 
                    indicator = True, validate = 'one_to_one', suffixes=['_planificacion', '_implementacion'])
    # segundo: calcular merge sólo para acciones unicamente "en implementación" (right merge + seleccionar 'right_only')
    data_rigt_merge = pd.merge(data_plani.drop(columns=cols_duplicadas),data_impl, on = cols_match, how = 'right', 
                               indicator = True, validate = 'one_to_one', suffixes=['_planificacion', '_implementacion'])
    data_rigt_merge_full = data_rigt_merge[data_rigt_merge._merge == 'right_only']    
    # tercero: concatenar
    data_tot = pd.concat([data,data_rigt_merge_full])
    
    # cuarto: para los caso duplicados en bbddd concatenada (todos del tipo "valor_invalido") considerar descripción como nombre acción
    data_tot.loc[data_tot.duplicated(['agno','rbd','Nombre Actividad'], keep = False),'Nombre Actividad'] = \
    data_tot.loc[data_tot.duplicated(['agno','rbd','Nombre Actividad'], keep = False),'Descripción del Plan'] 
    
    data_tot['fila_duplicada_merged'] = data_tot.duplicated(['agno','rbd','Nombre Actividad'], keep=False).astype(int)
    
    
    # Guardar en diccionarios filas duplicadas 
    dict_duplicados_post['duplicados_{}'.format(a)] = data_tot[data_tot.fila_duplicada_merged>0].copy()
    
    
    # quinto: botar los que siguen duplicados excepto uno (todos del tipo "valor_invalido")
    data_tot = data_tot.drop_duplicates( subset = ['agno','rbd','Nombre Actividad'], keep = 'first')

    
    # Rellenar columnas "EN" para uso en creacion de index 
    data_tot.En_planificacion_anual.fillna(0, inplace=True)
    data_tot.En_reporte_implementacion.fillna(0, inplace=True)
    
    # Crear index: rbd + año + counter + en planificacion + en implementacion
    data_tot['id_accion'] =  data_tot.rbd.astype(str) + data_tot.agno.astype(str) + \
                         (data_tot.groupby(['rbd']).cumcount()+1).astype(str) + \
                         data_tot.En_planificacion_anual.astype(int).astype(str) + \
                         data_tot.En_reporte_implementacion.astype(int).astype(str)
    
    # mejorar formato de variables con contenido de texto
    for col in content_cols + ['Nombre Actividad norm']:
        # pasar a minúsculas
        data_tot[col] = data_tot[col].str.lower()
        data_tot[col] = data_tot[col].astype(str)
        # remover espacios al inicio y final 
        data_tot[col] = data_tot[col].str.strip()
    
    # coordinar formatos montos 
    for col in ['monto_subvencion_general','monto_sep', 'monto_pie', 'monto_mantenimiento', 'monto_pro_retencion',
           'monto_internado', 'monto_reforzamiento', 'monto_feap', 'monto_aporte_municipal', 'monto_total']:
        if data_tot[col].dtype == 'object':
            data_tot[col] = data_tot[col].str.replace(',','')
            data_tot[col] = pd.to_numeric(data_tot[col])
            
    # Consolidad todos los años 
    if a == agno_inicial:
        df_planificacion_implementacion_total = data_tot.copy()
    else:
        df_planificacion_implementacion_total = pd.concat([df_planificacion_implementacion_total,data_tot])
    
```

```python
n = 0
for root, dirs, files in os.walk(r'Datos\Iniciales\Adicionales\SIMCE'):
    for file in files:
            if ('_rbd' in file) and ('idps' not in file) and (('csv' in file) | ('txt' in file)):
                print(file)
                path_file = os.path.join(root, file)
                if 'simce4b2018_rbd_publica_final' in file:
                    data = data = pd.read_csv(path_file,sep = '	', encoding = 'unicode_escape')
                else:
                    data = pd.read_csv(path_file, sep='|',encoding = 'unicode_escape')
                    
                mat = [x for x in data.columns if 'prom_mate' in x][0]
                lect = [x for x in data.columns if 'prom_lect' in x][0]
                
                data = data.rename(columns = {mat: 'prom_mate', lect :'prom_lect'})
                cols = ['agno', 'rbd'] + [x for x in data.columns if 'prom_lect' in x or 'prom_mate' in x]
        
                if n==0:
                    df_simce_tot = data[cols].copy()
                else:
                    df_simce_tot = pd.concat([df_simce_tot,data[cols]])
                n+=1

df_simce_tot = df_simce_tot.groupby(['agno','rbd'],as_index=False).mean()
for col in ['prom_mate','prom_lect']:
    df_simce_tot[col] = df_simce_tot[col].replace({0:np.nan})
    
# promedio
df_simce_tot['prom_mate_lect'] = df_simce_tot[['prom_mate','prom_lect']].mean(axis=1)
```
