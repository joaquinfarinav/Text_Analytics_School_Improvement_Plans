```python
# cargar bbdd
df_accion = pd.read_csv(r'Datos\Finales\Nivel_accion\df_planificacion_implementacion_total_2018_2022_ACCIONES_FINAL.csv', low_memory=False)
```

```python
lista_topics = list(df_accion.TEMA_palabras_clave.unique())
lista_ordenada = []
for x in lista_topics:
    if str(x) != 'nan':
        x_split = x.split('_')
        lista_ordenada += [[x_split[0], x_split[1:]]]
        
pd.DataFrame(lista_ordenada).sort_values(by=[0]).to_excel('diccionario_topic_modeling.xlsx')
```

```python
j = 0
for c in df_accion.columns:
    print(j, c)
    j+=1
```

```python
df_accion.en_reporte_implementacion.fillna(0, inplace=True)
df_accion.en_planificacion_anual.fillna(0, inplace=True)
```

```python
for col in df_accion.columns:
    for char in [':', '.', '\n', '(',')']:
        if char in col:
            print(col)
            df_accion = df_accion.rename(columns = {col:col.replace(char,'')})
            
    if col[-1] == '_':  
        col0 = col
        while col[-1] == '_':
            col = col[:-1]
        print(col)
        df_accion = df_accion.rename(columns = {col0:col})
```

```python
# Esquema de registro
dict_dict = {1: 'Unidad', 2:'Tema', 3:'Subtema'}
df_ER2 = pd.DataFrame(df_accion.columns, columns = ['Variable'])
df_ER2.loc[:,'Descripción'] = ''

for ind, row in df_ER2.iterrows():
    val = row.Variable
    i = 0
    for dicc in [dict_unidad,dict_tema,dict_subtema]:
        i+=1
        for x in dicc.items():
            if x[1] in val:
                df_ER2.loc[ind,'Descripción'] = dict_dict[i] +  ' : ' + x[0] + ' - concepto : ' + ' '.join(val.split('_')[3:])
                print(' '.join(val.split('_')[3:]))
```
